```python
for i in range(1,101):
    if(i%3==0 and i%5==0):
        print("FizzBuzz")
    elif(i%3==0):
        print("Fizz")
    elif(i%5==0):
        print("Buzz")
    else:
        print(i)
```

```python
p=0
q=1
list_f=[p,q]

while q<55:
    fibo = p+q
    list_f.append(fibo)
    
    p=q
    q=fibo
    
    
print(list_f)
```
