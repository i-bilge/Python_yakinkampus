```python
#import collections yerine sadece counter i aliyoruz.
from collections import Counter
import random
```

```python
list1 = random.sample(range(10), k=4)
list2 = random.sample(range(10), k=4)
list3 = random.sample(range(10), k=4)
list4 = random.sample(range(10), k=4)

liste_listesi = [list1, list2, list3, list4]
list_toplam = list1 + list2 + list3 + list4

print(liste_listesi)
print(list_toplam)
```

```python
for index, liste in enumerate(liste_listesi):
    print('{}. liste \t {}'.format(index+1, liste))
#simdi hangi elemandan kacar tane oldugunu istiyorum:
Counter(list_toplam)
```

```python
#noramalde cok satirli metinlerde her cünlenin sonuna \n koynak lazim
#diger yöntem de 3tirnak icerisine yazmaktir.
sarki = """
Yine bana gel
Yana yana yine beni sev
Yine bana gel
Yana yana yine beni sev"""
print(sarki)

#Cunter i uygularsak her hafri sorgular, her kelime icin önce splitle ayirir:
Counter(sarki.split())
```

```python
#kücük büyük harf farki olmasin diye de:
Counter(sarki.lower().split())#noktali emirleri de hep soldan saga uygular
```
