```python
#"https://image-charts.com/chart"
import requests
```

```python
grafik_URL = "https://image-charts.com/chart"

payload = {
    "chs": "700x190",
    "chd": "t:60,40",
    "cht": "p3",
    "chl": "Hello|World",
    "chan": None,
    "chf": "ps0-0,lg,45,ffeb3b,0.2,f44336,1|ps0-1,lg,45,8bc34a,0.2,009688,1"
}#buralardaki pipe(|) isareti veya anlaminda degil tam olarak.
response = requests.post(grafik_URL, data=payload)
```

```python
response.status_code #bu 200 calisiyor demek
```
